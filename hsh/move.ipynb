```python
# 필요한 라이브러리 및 모듈을 임포트

import shutil
import glob
import os
```

```python
# 경로 지정

train_images = "/Users/SBA/Documents/GitHub/3monkey_yolo/hsh/datasets/train/images/"
train_labels = "/Users/SBA/Documents/GitHub/3monkey_yolo/hsh/datasets/train/labels/"
val_images = "/Users/SBA/Documents/GitHub/3monkey_yolo/hsh/datasets/val/images/"
val_labels = "/Users/SBA/Documents/GitHub/3monkey_yolo/hsh/datasets/val/labels/"
```

```python
# 이동할 디렉토리 패턴

directory_pattern_train = os.path.join(train_images, "Suwon_CH0[1-4]*")
directory_pattern_val = os.path.join(val_images, "Suwon_CH0[1-4]*")
directory_pattern_val_labels = os.path.join(val_labels, "Suwon_CH0[1-4]*")
directory_pattern_train_labels = os.path.join(train_labels, "Suwon_CH0[1-4]*")
```

```python
try:
    # 이동할 디렉토리 목록 가져오기
    directories_to_move_train = glob.glob(directory_pattern_train)
    directories_to_move_val = glob.glob(directory_pattern_val)
    directories_to_move_val_labels = glob.glob(directory_pattern_val_labels)
    directories_to_move_train_labels = glob.glob(directory_pattern_train_labels)

    # 각 디렉토리의 파일을 이동할 디렉토리로 복사
    for source_directory in directories_to_move_train:
        source_files = os.listdir(source_directory)
        for file_name in source_files:
            source_path = os.path.join(source_directory, file_name)
            destination_path = os.path.join(train_images, file_name)
            shutil.move(source_path, destination_path)

    for source_directory in directories_to_move_val:
        source_files = os.listdir(source_directory)
        for file_name in source_files:
            source_path = os.path.join(source_directory, file_name)
            destination_path = os.path.join(val_images, file_name)
            shutil.move(source_path, destination_path)

    for source_directory in directories_to_move_val_labels:
        source_files = os.listdir(source_directory)
        for file_name in source_files:
            source_path = os.path.join(source_directory, file_name)
            destination_path = os.path.join(val_labels, file_name)
            shutil.move(source_path, destination_path)

    for source_directory in directories_to_move_train_labels:
        source_files = os.listdir(source_directory)
        for file_name in source_files:
            source_path = os.path.join(source_directory, file_name)
            destination_path = os.path.join(train_labels, file_name)
            shutil.move(source_path, destination_path)

    # 모든 디렉토리 삭제
    all_directories_train = glob.glob(os.path.join(train_images, "*/"))
    all_directories_val = glob.glob(os.path.join(val_images, "*/"))
    all_directories_val_labels = glob.glob(os.path.join(val_labels, "*/"))
    all_directories_train_labels = glob.glob(os.path.join(train_labels, "*/"))

    for directory in all_directories_train:
        shutil.rmtree(directory)

    for directory in all_directories_val:
        shutil.rmtree(directory)

    for directory in all_directories_val_labels:
        shutil.rmtree(directory)

    for directory in all_directories_train_labels:
        shutil.rmtree(directory)

except Exception as e:
    print(f"Error: {e}")
```
