```python
# 필요한 라이브러리 및 모듈을 임포트

import shutil
import glob
import os
```

```python
# 경로 지정

train_images_src_pattern = "/home/jupyter/datasets/train/images/Suwon_CH{}*/"
val_images_src_pattern = "/home/jupyter/datasets/val/images/*/"

val_labels_src_pattern = "/home/jupyter/datasets/val/labels/*/*"

train_labels_src_pattern = "/home/jupyter/datasets/train/labels/Suwon_CH{}*/"

train_images_dest = "/home/jupyter/datasets/train/images/"
val_images_dest = "/home/jupyter/datasets/val/images/"

val_labels_dest = "/home/jupyter/datasets/val/labels/"
train_labels_dest = "/home/jupyter/datasets/train/labels/"
```

```python
# Train 데이터셋의 이미지 파일을 이동 및 삭제

for ch_num in range(1, 5):
    ch_pattern = f"Suwon_CH{ch_num:02d}*"

    for src_path in glob.glob(train_images_src_pattern.format(ch_num) + ch_pattern):
        shutil.move(src_path, train_images_dest)

shutil.rmtree("/home/jupyter/datasets/train/images/*/")
```

```python
# Validation 데이터셋의 이미지 파일을 이동 및 삭제

for ch_num in range(1, 5):
    ch_pattern = f"Suwon_CH{ch_num:02d}*"

    for src_path in glob.glob(val_images_src_pattern + ch_pattern):
        shutil.move(src_path, val_images_dest)

shutil.rmtree("/home/jupyter/datasets/val/images/*/")
```

```python
# Validation 데이터셋의 라벨 파일을 이동 및 삭제

for src_path in glob.glob(val_labels_src_pattern):
    shutil.move(src_path, val_labels_dest)

shutil.rmtree("/home/jupyter/datasets/val/labels/*/")
```

```python
# Train 데이터셋의 라벨 파일을 이동 및 삭제

for ch_num in range(1, 5):
    ch_pattern = f"Suwon_CH{ch_num:02d}*"

    for src_path in glob.glob(train_labels_src_pattern.format(ch_num) + ch_pattern):
        shutil.move(src_path, train_labels_dest)

shutil.rmtree("/home/jupyter/datasets/train/labels/*/")
```
